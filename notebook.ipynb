# Sample Data

```python
-- Explore the data in the table
SELECT *
FROM companies;
```

## Querying tables

Get city,country columns from the table and order by the country ascending order

```python
--Get city,country columns from the table and order by the country ascending order
SELECT city,country
FROM companies
ORDER BY country ASC;
```

Get the first 10 rows from the companies list and want them have acceding ordder for company_id

```python
--Get the first 10 rows from the companies list and want them have acceding ordder for company_id
SELECT *
FROM companies
ORDER BY company_id ASC
LIMIT 10;
```

Get a unique list of countries where there are listings

```python
-- Get a unique list of countries where there are listings
SELECT DISTINCT country
FROM companies;
```

## Filtering on numeric columns

Get all the listings where country is china

```python
SELECT *
FROM companies
WHERE country = 'China';
```

Get all the listings where country is in Japan and Chile

```python
SELECT *
FROM companies
WHERE country IN( 'Japan','Chile')
```

Get all the listings where city starts with 'B' and where it does not end with 'n'

```python
SELECT *
FROM companies
WHERE city LIKE'B%' AND city NOT LIKE '%n';
```

Get all the listings for company_id is less than 500 and bigger than 58 , make the list order by country accesing

```python
SELECT *
FROM companies
WHERE company_id > 58 AND company_id < 500 
ORDER BY country ASC;
```

Filtering on multiple columns with text and numbers

Get all the listings in 'China' where the company_id is less than 500 

```python
SELECT *
FROM companies
WHERE company_id < 500 AND country ='China';
```

## Filtering on missing data

get all the listings where city is missing

```python
SELECT *
FROM companies
WHERE city = '';
```

get all the listings where city is NOT missing

```python
SELECT *
FROM companies
WHERE city != '';
```

## Simple aggregations

Get the total number of cities across all listings 

```python
SELECT COUNT(DISTINCT city) AS total_cities
FROM companies;
```

Get the total number of countries across all listings

```python
SELECT COUNT(DISTINCT country) AS total_countries
FROM companies;
```

Get the total number of continents across all listings

```python
SELECT COUNT(DISTINCT continent) AS total_countients
FROM companies;
```

Getting the highest company_id

```python
SELECT MAX(company_id) AS max_company_id
FROM companies;
```

Getting the lowest company_id

```python
SELECT MIN(company_id) AS min_company_id
FROM companies;
```

Getting the total number of companies for each country

```python
SELECT country, COUNT(company) AS total_companies
FROM companies
GROUP BY country 
ORDER BY total_companies DESC;
```

Getting the total number of companies for cities in the United States

```python
SELECT city, COUNT(company) AS total_companies
FROM companies
WHERE country = 'United States'
GROUP BY city
ORDER BY total_companies DESC;
```
